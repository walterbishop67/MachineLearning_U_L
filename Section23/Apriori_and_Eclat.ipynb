```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori

df = pd.read_csv(r"C:\Users\umut3\PycharmProjects\Apriori\Market_Basket_Optimisation.csv", header=None)
```

```python
transaction = []
for i in range(0, 7501):
    transaction.append([str(df.values[i, j]) for j in range(0, 20)])
```

```python
rules = apriori(transactions=transaction, min_support=0.003, min_confidence=0.2, min_lift=3, min_lenght=2, max_length=2)#min support 3*7(day)/75*1
```

```python
results = list(rules)
print(results)
```

```python
#Apriori
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    support = [result[1] for result in results]
    confidence = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, support, confidence, lifts))
```

```python
resultsinDataFrame = pd.DataFrame(inspect(results), columns=["left hand side", "right hand side", "support", "confidence", "lifts"])
resultsinDataFrame.nlargest(n=10, columns='lifts')
```

```python
#Eclat
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    support = [result[1] for result in results]
    return list(zip(lhs, rhs, support))
```

```python
resultsinDataFrame = pd.DataFrame(inspect(results), columns=["product1", "product2", "support"])
resultsinDataFrame.nlargest(n=10, columns='support')
```
